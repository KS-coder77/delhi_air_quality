```python
#import libraries 

import pandas as pd 
import numpy as np
import os, sys 
import sqlite3


import matplotlib.pyplot as plt
import numpy as np

import plotly as py
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as offline 
offline.init_notebook_mode(connected=True)

#load data 
india_air_data = pd.read_csv(r"C:\Users\Krupa\Documents\Krups Coding\Data Projects\Air quality data in India 2015-2020\city_day.csv")
india_air_data.head()
```

```python
missing_values_count = india_air_data.isnull().sum()
total_cells = np.product(india_air_data.shape)
total_missing = missing_values_count.sum()
percent_missing_vals = (total_missing/total_cells)*100
print(missing_values_count[:])
print('Overall percentage of missing values: ', percent_missing_vals)
```

```python
india_air_data = india_air_data.fillna(0)
india_air_data
```

```python
india_air_data['AQI_Bucket'] = india_air_data['AQI_Bucket'].replace(0, 'Other')
india_air_data
```

```python
print(india_air_data['City'].unique())
```

```python
ADB_data=india_air_data.loc[india_air_data['City'] == 'Ahmedabad', :]
ADB_data
```

```python
CHE_data=india_air_data.loc[india_air_data['City'] == 'Chennai', :]
CHE_data
```

```python
DEL_data=india_air_data.loc[india_air_data['City'] == 'Delhi', :]
DEL_data
```

```python
MUM_data=india_air_data.loc[india_air_data['City'] == 'Mumbai', :]
MUM_data
```

```python
india_air_data.describe()
```

```python
avg_readings = india_air_data.groupby('City').mean().reset_index()
avg_readings
```

```python
trace4=go.Bar(x=avg_readings['City'], y=avg_readings['CO'], marker_color=px.colors.qualitative.Dark24[0])
d4=[trace4]
layout=go.Layout(title='Average CO readings for Indian Cities 2015 - 2020', xaxis=dict(title='City'), yaxis=dict(title='CO Reading'))
fig=go.Figure(data=d4, layout=layout)
fig.show()
```

```python
trace0=go.Scatter(x=MUM_data['Date'], y=MUM_data['PM2.5'], name='Mumbai', line=dict(color=('red'),
          width=2))
```

```python
trace1=go.Scatter(x=ADB_data['Date'], y=ADB_data['PM2.5'], name='Ahmedabad', line=dict(color=('green'),
          width=2))
```

```python
trace2=go.Scatter(x=CHE_data['Date'], y=CHE_data['PM2.5'], name='Chennai', line=dict(color=('blue'),
          width=2))
```

```python
trace3=go.Scatter(x=DEL_data['Date'], y=DEL_data['PM2.5'], name='Delhi', line=dict(color=('yellow'),
          width=2))
```

```python
data=[trace0, trace1, trace2, trace3]
```

```python
layout=go.Layout(dict(title='PM2.5 readings for Indian Cities 2015 - 2020'), xaxis=dict(title='Date'), yaxis=dict(title='Readings'))
```

```python
fig=go.Figure(data=data, layout=layout)
#offline.iplot(fig)
fig.show()
```
