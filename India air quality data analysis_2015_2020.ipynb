```python
#Data analysis - average annual readings by city 
city_annual_avg=india_air_data.copy()
city_annual_avg['Date'] = pd.to_datetime(city_annual_avg['Date'])
city_annual_avg=city_annual_avg.groupby(['City', city_annual_avg.Date.dt.year]).mean().reset_index()
city_annual_avg.head(10)
```

```python
#Data correlations
city_annual_avg.corr().style.background_gradient(cmap='Greys')
```

```python
#r-values greater than 0.7 indicate a strong correlation. Therefore, it would be reasonable
#to saay there is a strong correlation between the following:-
# 1. AQI : PM2.5
# 2. NOx : NO
# 3. AQI : NO2
# 4. SO2 : CO
# 5. AQI : CO
# 6. AQI : SO2
```

```python
#Let's narrow in on two pollutants in particular, namely PM2.5 and NO2. 
#4. How has the data changed over time? - more readings from more cities?, less readings 
#from other cities?, other readings recorded? - compare pollutant readings over time 

fig=px.bar(city_annual_avg, x='City', y='PM2.5' , color='Date', title='Annual Average PM2.5 Readings for each City')
fig.show()
```

```python
#From the bar chart it is clear that the dataset for PM2.5 has expanded over the years to include more cities. Delhi, Lucknow
#and Patna have maintained a steady level of PM2.5. However, it is fair to say that most cities had a lower reading in 2020 -
#this could be due to the Pandemic with people working from home etc.
```

```python
fig=px.bar(city_annual_avg, x='City', y='NO2' , color='Date', title='Annual Average NO2 Readings for each City')
fig.show()
```

```python
#Similar to the PM2.5 readings, the NO2 dataset has also expanded over time to include more cities. Ahmedabad appears to 
#have high readings for 2018 and 2019, whereas the readings for Delhi are steadily decreasing over the years. This could be 
#due to increased measures to try to reduce the pollution in the city. 
```
